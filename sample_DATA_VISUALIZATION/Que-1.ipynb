```python
# Name :Parmar Sanjay
# Roll no:82
# Subject : Data Visualization
# Assignment : 2

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("employee_data.csv")

print("First 5 rows of dataset:")
print(df.head())

# 1. Average Salary of each Position
print("\nAverage Salary of each Position:")
avg_salary_position = df.groupby("Position")["Salary"].mean()
print(avg_salary_position)

plt.figure(figsize=(10, 6))
avg_salary_position.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Salary by Position")
plt.xlabel("Position")
plt.ylabel("Average Salary")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 2. Gender Distribution
print("\nTotal number of Male and Female employees:")
gender_count = df["Gender"].value_counts()
print(gender_count)

plt.figure(figsize=(6, 6))
gender_count.plot(kind="pie", autopct="%1.1f%%", colors=["#66b3ff", "#ff9999"], startangle=90)
plt.title("Gender Distribution in Employees")
plt.ylabel("")
plt.show()

# 3. Salary of Employees with 10-15 Years Experience
print("\nEmployees with 10-15 Years Experience and their Salary:")
exp_salary = df[(df["Experience (Years)"] >= 10) & (df["Experience (Years)"] <= 15)]
print(exp_salary[["ID", "Position", "Salary"]])

plt.figure(figsize=(10, 6))
plt.bar(exp_salary["ID"].astype(str), exp_salary["Salary"], color="purple", edgecolor="black")
plt.title("Salary of Employees with 10-15 Years Experience")
plt.xlabel("Employee ID")
plt.ylabel("Salary")
plt.xticks([], [])  # Hide IDs if too many
plt.show()

# 4. Number of Positions in Company
print("\nNumber of Employees in each Position:")
position_count = df["Position"].value_counts()
print(position_count)

plt.figure(figsize=(10, 6))
position_count.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Number of Positions in Company")
plt.xlabel("Position")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.show()

# 5. Best Position in terms of Salary
best_position = avg_salary_position.idxmax()
best_salary = avg_salary_position.max()
print(f"\nThe best position in terms of salary is {best_position} with an average salary of {best_salary:.2f}")

# 6. Additional Visualization: Salary Distribution (Boxplot)
plt.figure(figsize=(12, 6))
df.boxplot(column="Salary", by="Position", vert=False, grid=False)
plt.title("Salary Distribution by Position")
plt.suptitle("")
plt.xlabel("Salary")
plt.ylabel("Position")
plt.show()
```
